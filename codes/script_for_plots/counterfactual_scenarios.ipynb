```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
import os
```

```python
def compute_delta_death(confile1, confile2, path):
    # 2 is the one removing vaccine
    death_sim=np.load(path+confile1)["arr_0"]
    death_sim2=np.load(path+confile2)["arr_0"]
    delta_death=[]
    perc_delta_death=[]
    for i in range(len(death_sim)):
        dd=death_sim2[i]-death_sim[i]
        delta_death.append(dd)
        perc_delta_death.append(sum(dd)/sum(death_sim2[i])*100)
    return delta_death, perc_delta_death
```

```python
def compute_delta_death_daily(confile1, confile2, path):
    # 2 is the one removing vaccine
    death_sim=np.load(path+confile1)["arr_0"]
    death_sim2=np.load(path+confile2)["arr_0"]
    delta_death=[]
    perc_delta_death=[]
    for i in range(len(death_sim)):
        dd=death_sim2[i]-death_sim[i]
        delta_death.append(dd)
        perc_delta_death.append(np.where(death_sim2[i]!= 0, dd/death_sim2[i], 0))
    return delta_death, perc_delta_death
```

```python
def calculate_custom_box_stats(data):
    stats = []
    for col in data:  # 每个地区的数据
        q5 = np.percentile(col, 5)
        q50 = np.percentile(col, 50)
        q95 = np.percentile(col, 95)
        iqr = q95 - q5
        stats.append({
            'med': q50,     # 中位数
            'q1': q5,       # 5% 分位数
            'q3': q95,      # 95% 分位数
            'whislo': np.min(col),  # 下须
            'whishi': np.max(col),  # 上须
            'fliers': []    # 离群点
        })
    return stats
```
