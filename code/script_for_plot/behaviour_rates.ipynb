```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bx(x, para):
    return 1-np.exp(-x*para)
```

```python
alphas = [10, 5, 3, 1, 0.5, 0.1, 0.01, 0.001, 0.0001]
```

```python
gammas= [10, 5, 3, 1, 0.5, 0.1, 0.01, 0.001, 0.0001]
```

```python
colors=["#3d4748","#475e64","#4e757e","#578e9b","#5da7ba","#62bed9","#b2dfeb","#d7edf3","#ecf7f8"]#["#bfc9ff","#8aa6ff","#669aff","#67c7ff","#66f2ff","#95faff","#c5f3ff"]
colors2=["#536967","#5b8175","#5c9782","#5eb591","#68ceab","#71ddb7","#b0ead7","#ceece2","#deece6"]
labels1 = [r"$\alpha$=10", r"$\alpha$=5", r"$\alpha$=3", r"$\alpha$=1", r"$\alpha$=0.5", r"$\alpha$=0.1", r"$\alpha$=0.01", r"$\alpha$=0.001", r"$\alpha$=0.0001"]
labels2=[r"$\gamma$=10", r"$\gamma$=5", r"$\gamma$=3", r"$\gamma$=1", r"$\gamma$=0.5", r"$\gamma$=0.1", r"$\gamma$=0.01", r"$\gamma$=0.001", r"$\gamma$=0.0001"]
fig, ax = plt.subplots(ncols=2,dpi=200, figsize=(16,4))
i = 0
for alpha in alphas:
    v = np.linspace(0, 1, 1000)
    ax[0].plot(v, bx(v, alpha), color=colors[i], linewidth=2.5, label=labels1[i])
    i += 1
i = 0
for gamma in gammas:
    v = np.linspace(0, 2.5, 1000)
    ax[1].plot(v, bx(v, gamma), color=colors2[i], linewidth=2.5, label=labels2[i])
    i += 1

# Adjust the position of the annotations and titles
ax[0].annotate("Behavioural transition rates", xy=(1.22, 1.2), xycoords='axes fraction', rotation=0, fontsize=20, va='center', ha='center')
ax[0].annotate("A", xy=(0., 1.04), xycoords='axes fraction', fontsize=20, va='center', ha='left', weight="bold")
ax[0].set_title("Transition rate from NC to C", fontsize=20)
ax[1].annotate("B", xy=(0., 1.04), xycoords='axes fraction', fontsize=20, va='center', ha='left', weight="bold")
ax[1].set_title("Transition rate from C to NC", fontsize=20)

# Set labels for axes
ax[0].set_xlabel('fraction of vaccinated individuals', fontsize=16)
ax[1].set_xlabel('daily new deaths per 100,000 person', fontsize=16)
ax[0].set_ylabel(r'h($\alpha$, $frac_V$)', fontsize=18, rotation=90, labelpad=15)
ax[1].set_ylabel(r'h($\gamma$, $rate_D$)', fontsize=18, rotation=90, labelpad=10)

# Add legends outside the plot
ax[0].legend(frameon=False, fontsize=16, loc='center left', bbox_to_anchor=(1, 0.5))
ax[1].legend(frameon=False, fontsize=16, loc='center left', bbox_to_anchor=(1, 0.5))

# Add gridlines
ax[0].grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
ax[1].grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")

plt.subplots_adjust(wspace=0.5)
# Save the figure
plt.savefig(r'.\figures\behavioural_rates.pdf', bbox_inches="tight")
```
