```python
import pickle as pkl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tabulate import tabulate
```

```python
def get_nonzero_startpoint6_10(region, path):
    files = [f for f in os.listdir(path) if region in f and 'npz' in f]
    #files = [file for file in files if "_06_" in file or "_07_" in file or "_08_" in file or "_09_" in file or "_10_" in file]
    start_points=[]
    for file in files:
        death_sim=np.load(path+'\\'+file)["arr_0"]
        sim_md = np.median(death_sim, axis=0)
        for i in range(0, len(sim_md)):
            if sim_md[i]>0:
                start_points.append(i)
                break
    # print(start_points)
    # print(max(start_points))
    return max(start_points)
```

```python
def avg_wmape(region, confile, start_yw, end_yw, start_point, path, cut_n):
    df=pd.read_csv("../data/regions/" + region + "/epidemic/deaths.csv")
    death_real=df.loc[(df["year_week"] >= start_yw) & \
                                             (df["year_week"] <= end_yw)]["deaths"].values
    death_real=death_real[start_point:]
    death_sim=np.load(path+"\\"+confile)["arr_0"]
    #death_sim=np.load(f"{path}\{confile}")["arr_0"]
    md_sim=np.median(death_sim, axis=0)
    if cut_n==0:
        md_sim=md_sim[start_point:]
    else:
        md_sim=md_sim[start_point:][:-cut_n]
        death_real=death_real[:-cut_n]
    #error=0
    error = np.sum(np.abs(death_real - md_sim)) / np.sum(np.abs(death_real))
    return error
    #for i in range(len(death_sim)):
    #    error+=np.sum(np.abs(death_real - death_sim[i])) / np.sum(np.abs(death_real))
    #return error/len(death_sim)
```

```python
def rmse(region, confile, start_yw, end_yw, start_point, path, cut_n):
    df=pd.read_csv("../data/regions/" + region + "/epidemic/deaths.csv")
    death_real=df.loc[(df["year_week"] >= start_yw) & \
                                             (df["year_week"] <= end_yw)]["deaths"].values
    death_real=death_real[start_point:]
    death_sim=np.load(path+"\\"+confile)["arr_0"]
    #death_sim=np.load(f"{path}\{confile}")["arr_0"]
    md_sim=np.median(death_sim, axis=0)
    if cut_n==0:
        md_sim=md_sim[start_point:]
    else:
        md_sim=md_sim[start_point:][:-cut_n]
        death_real=death_real[:-cut_n]
    
    squared_diff = (death_real - md_sim) ** 2
    mean_squared_diff = np.mean(squared_diff)
    rmse_value = np.sqrt(mean_squared_diff)

    # print('nlogssr',len(death_real)*np.log(squared_diff))
    # print('rmse',rmse_value)
    return rmse_value
```

```python
def AIC(region, k, confile, start_yw, end_yw, start_point, path, cut_n):
    k=k
    df=pd.read_csv("../data/regions/" + region + "/epidemic/deaths.csv")
    death_real=df.loc[(df["year_week"] >= start_yw) & \
                                             (df["year_week"] <= end_yw)]["deaths"].values
    death_real=death_real[start_point:]
    death_sim=np.load(path+"\\"+confile)["arr_0"]
    md_sim=np.median(death_sim, axis=0)[start_point:]
    if cut_n!=0:
        md_sim=md_sim[:-cut_n]
        death_real=death_real[:-cut_n]
    # ssr=0
    # for j in range(n):
    #     ssr=ssr+(md_sim[j]-death_real[j])*(md_sim[j]-death_real[j]) # j不是number of trajectories，而是一条trajectory所有的数据点，只对比median trajectory和real_deaths
    # ssr=ssr/n
    n = len(death_real)
    ssr = np.mean((np.array(md_sim) - np.array(death_real)) ** 2)
    # error = np.sum(np.abs(death_real - md_sim)) / np.sum(np.abs(death_real))
    return n*np.log(ssr) + 2*k        #+2*k*(k+1)/(n-k-1)
```

```python
def BIC(region, k, confile, start_yw, end_yw, start_point, path, cut_n):
    k = k
    
    df = pd.read_csv("../data/regions/" + region + "/epidemic/deaths.csv")
    death_real = df.loc[(df["year_week"] >= start_yw) & \
                       (df["year_week"] <= end_yw)]["deaths"].values
    death_real = death_real[start_point:]   
    death_sim=np.load(path+"\\"+confile)["arr_0"]
    md_sim=np.median(death_sim, axis=0)[start_point:]
    if cut_n!=0:
        md_sim=md_sim[:-cut_n]
        death_real=death_real[:-cut_n]
    # 计算残差平方和
    # ssr = 0
    # for j in range(n):
    #     ssr = ssr + (md_sim[j]-death_real[j])*(md_sim[j]-death_real[j])
    # ssr = ssr/n
    n = len(death_real)
    ssr = np.mean((np.array(md_sim) - np.array(death_real)) ** 2)
    return n*np.log(ssr) + k*np.log(n)
```

```python
import os
def get_6_10_aic_new(region, start_yw, end_yw, path, cut_n):
    # # n is the number of the last weeks cut off, not included in computation of AIC or BIC
    # print(region)
    aics=[]
    wmapes=[]
    #mapes=[]
    rmses=[]
    start_point = get_nonzero_startpoint6_10(region, path)
    all_files  = [f for f in os.listdir(path) if region in f and 'npz' in f]
    files=all_files
    #files  = [i for i in all_files if ("_06_" in i) or ("_07_" in i) or ("_08_" in i) or ("_09_" in i) or ("_10_" in i)]
    for i in files:
        if "_06_" in i:
            if "London" in i:
                aics.append(AIC(region, 5, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Lombardy" in i or "British Columbia" in i:
                aics.append(AIC(region, 6, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Sao Paulo" in i: 
                aics.append(AIC(region, 7, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
        else:
            if "London" in i:
                aics.append(AIC(region, 8, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Lombardy" in i or "British Columbia" in i:
                aics.append(AIC(region, 9, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Sao Paulo" in i: 
                aics.append(AIC(region, 10, i, start_yw, end_yw, start_point, path, cut_n))
                wmapes.append(avg_wmape(region, i, start_yw, end_yw, start_point, path, cut_n))
                rmses.append(rmse(region, i, start_yw, end_yw, start_point, path, cut_n))
                #mapes.append(mape(region, i, start_yw, end_yw, start_point, path, cut_n))
    return aics, wmapes, rmses
```

```python
import os
def get_6_10_bic_new(region, start_yw, end_yw, path, cut_n):
    # print(region)
    aics=[]
    wmapes=[]
    mapes=[]
    start_point = get_nonzero_startpoint6_10(region, path)
    all_files  = [f for f in os.listdir(path) if region in f and 'npz' in f]
    files=all_files
    # files  = [i for i in all_files if ("_06_" in i) or ("_07_" in i) or ("_08_" in i) or ("_09_" in i) or ("_10_" in i)]
    for i in files:
        if "_06_" in i:
            if "London" in i:
                aics.append(BIC(region, 5, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Lombardy" in i or "British Columbia" in i:
                aics.append(BIC(region, 6, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Sao Paulo" in i: 
                aics.append(BIC(region, 7, i, start_yw, end_yw, start_point, path, cut_n))
        else:
            if "London" in i:
                aics.append(BIC(region, 8, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Lombardy" in i or "British Columbia" in i:
                aics.append(BIC(region, 9, i, start_yw, end_yw, start_point, path, cut_n))
            elif "Sao Paulo" in i: 
                aics.append(BIC(region, 10, i, start_yw, end_yw, start_point, path, cut_n))
    # print("BICs:", aics)
    return aics
```

```python
import math
def AIC_weight(aics_list):
    smallest=min(aics_list)
    s=0
    for i in range(len(aics_list)):
        s+=math.exp(-(aics_list[i]-smallest)/2)
    ws=[]
    for i in range(len(aics_list)):
        ws.append(math.exp(-(aics_list[i]-smallest)/2)/s)
    return ws
```

```python
def compute_results(cut_n, subpath):
    path = f"..\\posteriors\\{subpath}\\"
    bc_aic, bc_wmape, bc_rmse = get_6_10_aic_new('British Columbia', '2020-42', '2021-26', path, cut_n)
    lom_aic, lom_wmape, lom_rmse = get_6_10_aic_new('Lombardy', '2020-41', '2021-26', path, cut_n)
    ld_aic, ld_wmape, ld_rmse = get_6_10_aic_new('London', '2020-47', '2021-26', path, cut_n)
    sp_aic, sp_wmape, sp_rmse = get_6_10_aic_new('Sao Paulo', '2021-03', '2021-39', path, cut_n)

    bc_bic = get_6_10_bic_new('British Columbia', '2020-42', '2021-26', path, cut_n)
    lom_bic = get_6_10_bic_new('Lombardy', '2020-41', '2021-26', path, cut_n)
    ld_bic = get_6_10_bic_new('London', '2020-47', '2021-26', path, cut_n)
    sp_bic = get_6_10_bic_new('Sao Paulo', '2021-03', '2021-39', path, cut_n)

    print('WMAPE\n')
    # wmapes
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 3) for i in bc_wmape])
    ls2=['Lombardy']
    ls2.extend([round(i, 3) for i in lom_wmape])
    ls3=['London']
    ls3.extend([round(i, 3) for i in ld_wmape])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 3) for i in sp_wmape])
    table = [title, ls1, ls2, ls3, ls4]
    
    print(tabulate(table, tablefmt="latex"))
    print((bc_wmape[0]-min(bc_wmape))/bc_wmape[0])
    print((lom_wmape[0]-min(lom_wmape))/lom_wmape[0])
    print((ld_wmape[0]-min(ld_wmape))/ld_wmape[0])
    print((sp_wmape[0]-min(sp_wmape))/sp_wmape[0])

    print('\n\nAIC\n')
    # aic
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 1) for i in bc_aic])
    ls2=['Lombardy']
    ls2.extend([round(i, 1) for i in lom_aic])
    ls3=['London']
    ls3.extend([round(i, 1) for i in ld_aic])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 1) for i in sp_aic])
    table = [title, ls1, ls2, ls3, ls4]
    
    print(tabulate(table, tablefmt="latex"))
    
    print('\n\nAIC weight\n')
    # AIC weights
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 2) for i in AIC_weight(bc_aic)])
    ls2=['Lombardy']
    ls2.extend([round(i, 2) for i in AIC_weight(lom_aic)])
    ls3=['London']
    ls3.extend([round(i, 2) for i in AIC_weight(ld_aic)])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 2) for i in AIC_weight(sp_aic)])
    table = [title, ls1, ls2, ls3, ls4]
    
    print(tabulate(table, tablefmt="latex"))

    print('\n\nBIC score\n')
    # bic
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 1) for i in bc_bic])
    ls2=['Lombardy']
    ls2.extend([round(i, 1) for i in lom_bic])
    ls3=['London']
    ls3.extend([round(i, 1) for i in ld_bic])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 1) for i in sp_bic])
    table = [title, ls1, ls2, ls3, ls4]
    
    print(tabulate(table, tablefmt="latex"))
    
    print('\n\nBIC weight\n')
    # BIC weights
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 2) for i in AIC_weight(bc_bic)])
    ls2=['Lombardy']
    ls2.extend([round(i, 2) for i in AIC_weight(lom_bic)])
    ls3=['London']
    ls3.extend([round(i, 2) for i in AIC_weight(ld_bic)])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 2) for i in AIC_weight(sp_bic)])
    table = [title, ls1, ls2, ls3, ls4]

    
    print(tabulate(table, tablefmt="latex"))

    print('\n\nRMSE\n')
    # rmse
    title=['', 'baseline', 'model 1', 'model 2', 'model 3', 'model 4']
    ls1=['British Columbia']
    ls1.extend([round(i, 3) for i in bc_rmse])
    ls2=['Lombardy']
    ls2.extend([round(i, 3) for i in lom_rmse])
    ls3=['London']
    ls3.extend([round(i, 3) for i in ld_rmse])      
    ls4=['Sao Paulo']
    ls4.extend([round(i, 3) for i in sp_rmse])
    table = [title, ls1, ls2, ls3, ls4]
    
    print(tabulate(table, tablefmt="latex"))

    print(AIC_weight(bc_aic))
    print(AIC_weight(bc_bic))

    print(AIC_weight(bc_aic)==AIC_weight(bc_bic))
```

```python
ncomp = 12
nage  = 16

fig, axes = plt.subplots(nrows=2, ncols=2, dpi=210, figsize=(12,7))
i, j = 0, 0

for region in ["British Columbia", "Lombardy", "London", "Sao Paulo"]:
    
    df=pd.read_csv("../data/regions/" + region + "/epidemic/deaths.csv")

    if region=="British Columbia":
        start_yw, end_yw = "2020-42", "2021-26"
    elif region=="Lombardy":
        start_yw, end_yw = "2020-41", "2021-26"
    if region=="London":
        start_yw, end_yw = "2020-47", "2021-26"
    if region=="Sao Paulo":
        start_yw, end_yw = "2021-03", "2021-39"
    
    death_real=df.loc[(df["year_week"] >= start_yw) & \
                                         (df["year_week"] <= end_yw)]["deaths"].values
    real_date=df.loc[(df["year_week"] >= start_yw) & \
                                         (df["year_week"] <= end_yw)]["year_week"].values

    colors = ['#6CC2BD', '#5A809E', '#F57D7C', '#c5c9c7', '#FEE4C4', \
      '#FFC1A6', '#acc2d9', '#c5c9c7', '#a8415b', '#b1d1fc']
    axes[i, j].grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    axes[i, j].spines['top'].set_visible(False)
    axes[i, j].spines['right'].set_visible(False)
    axes[i, j].minorticks_on()  
    
    path="..\\posteriors\\"
    files  = os.listdir(path)
    confile_list  = [i for i in files if (region in i) and ('.npz' in i)]
    confile_list  = [i for i in confile_list if ("_06_" in i) or ("_07_" in i) or ("_08_" in i) or ("_09_" in i) or ("_10_" in i)]

    #calculate deaths per 10,000 persons
    pop=pd.read_excel(f'../data/regions/{region}/demographic/pop_5years.xlsx')
    pop_total=pop['total'].sum()
    death_real = death_real/pop_total*100000

    
    p3=[0,0,0,0,0]
    p2=[0,0,0,0,0]
    p1=[0,0,0,0,0]

    # death_real=death_real[start:]
    # real_date=real_date[start:]
    for m in range(len(confile_list)):
        
        death_sim=np.load(path+"\\"+ confile_list[m])["arr_0"]
        death_sim = death_sim/pop_total*100000
        
        p1[m], = axes[i, j].plot(real_date, np.median(death_sim, axis=0), color=colors[m], alpha=1,linewidth=2)
        
        p2[m] = axes[i, j].fill_between(real_date, np.quantile(death_sim, axis=0, q=0.05), 
                             np.quantile(death_sim, axis=0, q=0.95), color=colors[m], alpha=0.3, linewidth=0)#6CC2BD
        p3[m] = axes[i, j].scatter(real_date,death_real,color="#5A809E", edgecolor="black",s=15,linewidth=0.2)
        
        

    array_index=[]
    for d in range(len(real_date)):
        if d%4==0:
            array_index.append(d)
    axes[i, j].set_xticks(array_index)
    labels = [real_date[d] for d in array_index]
    axes[i, j].set_xticklabels(labels, rotation=30)
    axes[i, j].tick_params(axis="y", labelsize=14)
    axes[i, j].grid(color='grey', linestyle='--', linewidth=0.35, alpha=0.6, axis="both")
    axes[i, j].legend(frameon=False)
    axes[i, j].set_title(region, fontsize=16)
    
    j += 1
    
    if j==2:
        j = 0
        i += 1
plt.subplots_adjust(hspace=0.45, wspace=0.25)

axes[1, 0].legend([(p1[0], p2[0]), (p1[1], p2[1]), (p1[2],p2[2]), (p1[3], p2[3]), (p1[4], p2[4]), p3[0]], 
                  ["baseline" , "model 1", "model 2", "model 3", "model 4", "real data"],bbox_to_anchor=[0.7, -1.25,1,1], frameon=False, ncol=3, fontsize=12)
axes[0, 0].set_ylabel("weekly death rate",fontsize=16, y=-0.1)
plt.savefig("./figures/comparison.pdf", bbox_inches="tight")
plt.show()
```
